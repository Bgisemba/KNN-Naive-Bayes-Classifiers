##OUtliers

```python
# dealing with outliers
# quantiles
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Removing outliers based on the IQR range 
df_out= df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

# Printing the shape of our new dataset
print(df_out.shape)

# Printing the shape of our old dataset
print(df.shape)
```

We decided not to drop the outliers since its a large percentage of the whole dataset and it would really affect out analysis.

#Modelling

```python
#viewing the columns of the dataset
df.columns
```

```python
#Separating the variables
X=df.drop(columns=['spam'],axis=1).values
y=df['spam'].values
```

```python
X
```

```python
y
```

Since our independent variables are continuos we will apply the Gaussian Naive Bayes Classifierto predict whether an email is a Spam or not.

##Gaussian Naive Bayes Classifier

##80:20

```python
# Splitting our data into a training set and a test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 

#Feature scaling
from sklearn.preprocessing import StandardScaler  
sc = StandardScaler()  
X_train = sc.fit_transform(X_train)  
X_test = sc.transform(X_test) 

#Training our model
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()  
model = clf.fit(X_train, y_train) 

#Predicting our test predictors
y_pred = model.predict(X_test)

# Evaluating the Algorithm 
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import metrics
print(f'Accuracy score =: {metrics.accuracy_score(y_test, y_pred)} ')
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

Spliting the data into **80:20** leads to an **accuracy score of 81%**.

**True Positive :**353

**True Negative :**331

**False Positive :** 142

**False Negative:**16

##Hyperparameter Tuning

We decided to carry out hyperparameter tuning to obtain the best parameters to apply in our model.

```python
#carrying out the Gris Search Cross Validation
from sklearn.model_selection import GridSearchCV
nb_classifier = GaussianNB()

params_NB = {'var_smoothing': np.logspace(0,-9, num=100)}
gs_NB = GridSearchCV(estimator=nb_classifier, 
                 param_grid=params_NB, 
                 cv=10,   
                 verbose=1, 
                 scoring='accuracy') 
gs_NB.fit(X_train, y_train)

gs_NB.best_params_
```

```python
gs_NB.best_estimator_
```
